# Test the pipeline (generator)
#### This notebook is to ensure the output of each function in pipeline.py meets the requirements.
### Load packages

```python
from pipeline import TrainGenerator,TestDataPreparation, ValidationDataPreparation
import time
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

## Test the pipeline without deformation

```python
myGenerator = TrainGenerator(batch_size=5,target_size = 704, deformation = False)
```

### To compare the time cost, record the time for generation without elastic deformation.

```python
time_start=time.time()#tic
data = next(myGenerator)
time_end=time.time()#toc
print('time cost',time_end-time_start,'s')
```

### Check the shape of outputs is (batch_size, width, height, channels).

```python
print(data[0].shape)#data
print(data[1].shape)#masks
```

### Check that the training data and training masks are agumented correctly and simultaneously.

```python
for i in range(data[1].shape[0]):
    plt.figure(figsize=(8,6))
    plt.subplot(121)
    img = np.squeeze(data[0][i])#Draw the framework to highlight the area for segmentation.
    u = (data[0].shape[1]-512)//2
    v = data[0].shape[1] - (data[0].shape[1]-512)//2
    cv2.line(img,(u,u),(u,v),(1,0,0),5)
    cv2.line(img,(u,u),(v,u),(1,0,0),5)
    cv2.line(img,(v,v),(u,v),(1,0,0),5)
    cv2.line(img,(v,v),(v,u),(1,0,0),5)
    plt.imshow(img,'gray')
    plt.subplot(122)
    plt.imshow(np.squeeze(data[1][i]),'gray')
    plt.show()
```
